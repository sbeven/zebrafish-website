```python

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Landscape.gene.expr_v2.csv")
df = df.iloc[:, :-2]

df
```

```python
df = df.rename(columns={s:s.replace("pct.", "") for s in df.columns})
df
```

```python
df = df.set_index("gene")
df.to_csv("Landscape.gene.expr.truncated.csv")
df
```

```python
collist = []
for name in df.columns:
    collist.append(name + "_spec")
collist
```

```python
df_spec = pd.DataFrame(columns=collist)
df_spec
```

```python
for column in df.columns:
    max = df.drop(column, axis=1).max(axis=1)
    df_spec[column + "_spec"] = 1 - max
df_spec

```

```python
df_spec.to_csv("Landscape_spec.csv")
```
